- **favorite_count**: Favorite count of the tweet.
- **timestamp_ms**: Time in milliseconds.

## 3.3 Master data file creation

Creating master data file needed extensive data processing and filtering steps. As the data size was huge and it consuming huge time, we used 4 different Jupyter notebook files to process our data for the master data file. It provides us a prominent way to deal with different data files and finally create a files with the expected features for master data file. We combined all the data based on time. For this project, we formed data in 2 min and 5 min groups. However, we find out 2min data group is better for our regression and prediction analysis as it allows us to explore more observations. Also, the cryptocurrency price varies a lot within minutes. Hence, 2min data can provide a standard overview of the relation between price and other predictors.

### 3.3.1 Tweets text processor for master data file

In this section we performed the following steps:
1. First we filter the data between 2018-02-26 02:55:00 and 2018-03-03 08:43:41 which is coherent with the price data. 

2. As the data was fetched from streaming API we didn't had any retweets count. To get the retweets count we search the keyword 'RT' (which stands for Re-Tweets) in the tweets and using this we count the number of retweets

3. We grouped the data by 2min and merged all the tweets text in that time group. We also add the retweets count in that time.

4. In this step we analyze the tweet text and extract the sentiment polarity score using textblob.

5. Finally we stored the data with columns <b>time (created_at), sentiment_score, no_tweets, and retweet_counts</b> in a file. 

The detailed codes are include in the jupyter notebook file named Tweets_text_processor_for_master_data_file.ipnyb in the data_processor folder. 

### 3.3.2 Tweets user processor for master data file

In this section we performed the following steps:
1. First we filter the user data between 2018-02-26 02:55:00 and 2018-03-03 08:43:41 which is coherent with the price data. The user data consists following columns. <b>['id', 'created_at', 'favorite_count', 'favorited', 'retweet_count','user_id', 'user_created_at', 'user_followers_count', 'user_friends_count', 'user_location', 'user_screen_name', 'user_verified', 'user_description', 'created_at_time'] </b>

2. We filtered and dropped the duplicate users. 

3. We extracted the age of the user ID (the age of the twitter account of a user) using exact tweet time and user id creation time. 

4. We grouped the data by 2min and find out the aggregate some of the different variables in that time (e.g. user_followers_count, user_friends_count). We also extracted the number of users and the number of verified users in that time. 

5. Finally we stored the data with columns <b>'created_at', 'user_followers_count', 'user_friends_count', 'user_verified', 'no_users', and 'avg_id_age_days'</b> in a file. 

The detailed codes are include in the jupyter notebook file named Tweets_user_processor_for_master_data_file.ipnyb in the data_processor folder. 

### 3.3.3 Price data processor for master data file

Processing price data involving the following steps: 

1. Read the data and filter the data data between 2018-02-26 02:55:00 and 2018-03-03 08:43:41. The price data consists <b>close, high, low, open, time, volumefrom, and volumeto columns.</b>

2. Extracting the transaction volume of every minutes by subtracting volumefrom from volumeto column.

3. Grouping the data by 2 min and calculate the average value of all the variables in this 2min. 

4. Finally we stored the data with columns <b>'time', 'close', 'high', 'low', and 'volume'</b> in a file.

The detailed codes are include in the jupyter notebook file named Price_data_processor_for_master_data_file.ipnyb in the data_processor folder.

### 3.3.4 Master data file creation

In this section, we combined the three data files (price data, tweets text data, and tweets user data) stored by above three steps and form the master data file. We merged all three files by time.
